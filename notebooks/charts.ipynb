# Wykresy zależności czasu przetwarzania wideo przez wybrane modele z uwzględeniem różnic liczbie pomijanych klatek

```python
import cv2
import matplotlib.pyplot as plt
```

```python
processing_times_1920x1088 = [148.97, 77.00, 53.14, 40.93, 32.77]
processing_times_1280x720 = [119.10, 61.38, 41.99, 32.46, 26.22]
processing_times_640x640 = [109.46, 56.88, 39.18, 35.13, 27.40]

acc_1920x1088 = [100, 100, 100, 100, 91]
acc_1280x720 = [100, 100, 100, 100, 91]
acc_620x620 = [82, 82, 72, 30.7, 27]

frame_strides = [1, 2, 3, 4, 5]
```

```python
plt.figure(figsize=(10, 6))
plt.axhline(y=280, color='red', linestyle='--', label='Video Duration (280s)')
plt.plot(frame_strides, processing_times_1920x1088, marker='o', label='1920x1088')
plt.plot(frame_strides, processing_times_1280x720, marker='o', label='1280x720')
plt.plot(frame_strides, processing_times_640x640, marker='o', label='640x640')
plt.xlabel('Frame Stride')
plt.ylabel('Processing Time [s]')
plt.title('15_04_25/7-8/Aerodynamika_fragment.mp4 - Yolo11s')
plt.xticks(frame_strides)
plt.grid(True)
plt.legend()
plt.savefig('../img/processing_time_vs_frame_stride_aerodynamika.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(frame_strides, acc_1920x1088, marker='o', label='1920x1088')
plt.plot(frame_strides, acc_1280x720, marker='o', label='1280x720')
plt.plot(frame_strides, acc_620x620, marker='o', label='640x640')
plt.xlabel('Frame Stride')
plt.ylabel('Accuracy [%]')
plt.title('15_04_25/7-8/Aerodynamika_fragment.mp4 - Yolo11s')
plt.xticks(frame_strides)
plt.grid(True)
plt.legend()
plt.savefig('../img/accuracy_vs_frame_stride_aerodynamika.png')
plt.show()
```
